```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import re

# --- Teks Contoh ---
text = "Saya Ghiraldy Patrick Toshka Anggawan, seorang mahasiswa Teknik Informatika yang sedang belajar Deep Learning di kota Sorong. Ini sangat menarik!"

print(f"Teks Asli:\n{text}\n")

# --- 1. Lowercasing ---
text_lower = text.lower()
print(f"1. Lowercasing:\n{text_lower}\n")

# --- 2. Tokenization ---
# Menghapus tanda baca terlebih dahulu agar tidak menjadi token sendiri (opsional)
text_no_punct = re.sub(r'[^\w\s]', '', text_lower)
tokens = word_tokenize(text_no_punct)
print(f"2. Tokenization (dan Hapus Tanda Baca):\n{tokens}\n")

# --- 3. Stop Word Removal ---
# Mendapatkan daftar stop words Bahasa Indonesia dari NLTK
list_stopwords = set(stopwords.words('indonesian'))
tokens_without_sw = [word for word in tokens if not word in list_stopwords]
print(f"3. Stop Word Removal:\n{tokens_without_sw}\n")

# --- 4. Stemming (Menggunakan Sastrawi) ---
# Membuat stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

tokens_stemmed = [stemmer.stem(word) for word in tokens_without_sw]
print(f"4. Stemming:\n{tokens_stemmed}\n")

print("--- Hasil Akhir (NLTK + Sastrawi) ---")
print(" ".join(tokens_stemmed))
```

```python
import numpy as np
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# --- Dataset Contoh ---
data = [
    ("Timnas Indonesia menang lawan Vietnam di kualifikasi piala dunia.", "Olahraga"),
    ("Bulu tangkis ganda putra raih medali emas.", "Olahraga"),
    ("PSSI umumkan pelatih baru untuk timnas.", "Olahraga"),
    ("Perlombaan lari maraton digelar di Jakarta.", "Olahraga"),
    ("Komite Olimpiade resmikan cabang olahraga baru.", "Olahraga"),
    ("Google rilis versi terbaru Android dengan fitur AI canggih.", "Teknologi"),
    ("Startup teknologi Indonesia dapat pendanaan besar.", "Teknologi"),
    ("Review laptop gaming terbaru dengan prosesor intel.", "Teknologi"),
    ("Bahaya kecanduan media sosial dibahas dalam seminar teknologi.", "Teknologi"),
    ("Apple perkenalkan iPhone model terbaru dengan chip A18.", "Teknologi"),
]

texts, labels = zip(*data)

# --- Fungsi Preprocessing ---
stop_words_indo = set(stopwords.words('indonesian'))

def preprocess_text(text):
    text = text.lower()  # Lowercasing
    text = re.sub(r'[^\w\s]', '', text)  # Hapus tanda baca
    tokens = word_tokenize(text)  # Tokenization
    # Hapus stop words & kata pendek (opsional)
    tokens = [word for word in tokens if word not in stop_words_indo and len(word) > 2]
    return tokens

# Preprocess semua teks
processed_texts_tokens = [preprocess_text(text) for text in texts]
processed_texts_string = [" ".join(tokens) for tokens in processed_texts_tokens]

# --- Membagi Data (Train & Test) ---
X_train_str, X_test_str, y_train, y_test, X_train_tok, X_test_tok = train_test_split(
    processed_texts_string, labels, processed_texts_tokens, test_size=0.3, random_state=42
)

print("Setup Selesai.")
print(f"Jumlah data latih: {len(X_train_str)}")
print(f"Jumlah data tes: {len(X_test_str)}\n")
```

```python
print("--- Klasifikasi dengan TF-IDF ---")

# --- Membuat Vektor TF-IDF ---
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train_str)
X_test_tfidf = vectorizer.transform(X_test_str)

# --- Melatih Model Klasifikasi ---
model_tfidf = LogisticRegression()
model_tfidf.fit(X_train_tfidf, y_train)

# --- Prediksi & Evaluasi ---
y_pred_tfidf = model_tfidf.predict(X_test_tfidf)

accuracy_tfidf = accuracy_score(y_test, y_pred_tfidf)
report_tfidf = classification_report(y_test, y_pred_tfidf)

print(f"Akurasi TF-IDF: {accuracy_tfidf:.2f}")
print("Laporan Klasifikasi TF-IDF:")
print(report_tfidf)
```

```python
print("\n--- Klasifikasi dengan Word2Vec ---")

# --- Melatih Model Word2Vec ---
# Ukuran vektor (vector_size) kecil karena data sedikit
w2v_model = Word2Vec(sentences=X_train_tok, vector_size=50, window=3, min_count=1, workers=4)
print("Model Word2Vec selesai dilatih.")

# --- Fungsi untuk Membuat Vektor Dokumen (Rata-rata Vektor Kata) ---
def create_document_vector(tokens, model):
    vector = np.zeros(model.vector_size)
    count = 0
    for word in tokens:
        if word in model.wv:  # Periksa apakah kata ada di vocabulary model
            vector += model.wv[word]
            count += 1
    if count != 0:
        vector /= count  # Rata-ratakan
    return vector

# --- Membuat Vektor Dokumen untuk Data Latih & Tes ---
X_train_w2v = np.array([create_document_vector(tokens, w2v_model) for tokens in X_train_tok])
X_test_w2v = np.array([create_document_vector(tokens, w2v_model) for tokens in X_test_tok])

# --- Melatih Model Klasifikasi ---
# Gunakan solver yang berbeda jika ada warning, atau scale data jika perlu
model_w2v = LogisticRegression()
model_w2v.fit(X_train_w2v, y_train)

# --- Prediksi & Evaluasi ---
y_pred_w2v = model_w2v.predict(X_test_w2v)

accuracy_w2v = accuracy_score(y_test, y_pred_w2v)
report_w2v = classification_report(y_test, y_pred_w2v)

print(f"\nAkurasi Word2Vec: {accuracy_w2v:.2f}")
print("Laporan Klasifikasi Word2Vec:")
print(report_w2v)

# --- Contoh Prediksi Baru ---
new_text_olahraga = "Pertandingan sepak bola berakhir imbang"
new_text_teknologi = "Ponsel lipat terbaru diluncurkan"

# Preprocess
new_tok_olahraga = preprocess_text(new_text_olahraga)
new_tok_teknologi = preprocess_text(new_text_teknologi)

# Vektor TF-IDF
new_tfidf_olahraga = vectorizer.transform([" ".join(new_tok_olahraga)])
new_tfidf_teknologi = vectorizer.transform([" ".join(new_tok_teknologi)])

# Vektor W2V
new_w2v_olahraga = create_document_vector(new_tok_olahraga, w2v_model).reshape(1, -1)
new_w2v_teknologi = create_document_vector(new_tok_teknologi, w2v_model).reshape(1, -1)


print("\nPrediksi Teks Baru:")
print(f"'{new_text_olahraga}' -> TF-IDF: {model_tfidf.predict(new_tfidf_olahraga)[0]}, Word2Vec: {model_w2v.predict(new_w2v_olahraga)[0]}")
print(f"'{new_text_teknologi}' -> TF-IDF: {model_tfidf.predict(new_tfidf_teknologi)[0]}, Word2Vec: {model_w2v.predict(new_w2v_teknologi)[0]}")
```

```python
from transformers import pipeline
import time

# --- 1. Muat Pipeline Klasifikasi Teks ---
# Kita tentukan tugasnya ('sentiment-analysis')
# dan model spesifik yang ingin digunakan.
# Jika model tidak ada secara lokal, pipeline akan mengunduhnya secara otomatis.
print("Memuat pipeline klasifikasi sentimen...")
classifier = pipeline(
    'sentiment-analysis',
    model='mdhugol/indonesia-bert-sentiment-classification'
)
print("Pipeline siap digunakan.\n")

# --- 2. Siapkan Teks Contoh ---
texts_to_classify = [
    "Saya sangat senang dengan pelayanan di restoran ini, makanannya luar biasa!",
    "Filmnya sangat mengecewakan, ceritanya datar dan membosankan.",
    "Hari ini cuaca di Sorong cukup cerah.",
    "Produk ini bekerja sesuai ekspektasi, tidak ada yang istimewa.",
    "Pengiriman barang sangat lambat, saya menunggu lebih dari seminggu.",
    "Timnas Indonesia bermain dengan semangat juang tinggi, saya bangga!"
]

# --- 3. Lakukan Klasifikasi ---
print("--- Mulai Klasifikasi Teks ---")
start_time = time.time()

results = classifier(texts_to_classify)

end_time = time.time()
print(f"Klasifikasi selesai dalam {end_time - start_time:.2f} detik.\n")

# --- 4. Tampilkan Hasil ---
print("--- Hasil Klasifikasi ---")
for text, result in zip(texts_to_classify, results):
    label = result['label']
    score = result['score']
    print(f"Teks  : '{text}'")
    print(f"Hasil : {label} (Skor: {score:.4f})\n")

# --- Contoh Tunggal ---
print("--- Contoh Klasifikasi Tunggal ---")
text_single = "Belajar Deep Learning ternyata sangat menantang tapi seru."
result_single = classifier(text_single)
print(f"Teks  : '{text_single}'")
print(f"Hasil : {result_single[0]['label']} (Skor: {result_single[0]['score']:.4f})")
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# --- 1. Pilih Model ---
# Kita gunakan Mistral 7B Instruct v0.2
model_id = "mistralai/Mistral-7B-Instruct-v0.2"

# --- 2. Muat Tokenizer ---
print(f"Memuat tokenizer untuk {model_id}...")
# Tokenizer mengubah teks -> angka (token ID) dan sebaliknya
tokenizer = AutoTokenizer.from_pretrained(model_id)

# Penting: Pastikan ada padding token jika belum ada (berguna untuk batching)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token # Gunakan end-of-sentence token sebagai pad

print("Tokenizer berhasil dimuat.")

# --- 3. Muat Model ---
print(f"Memuat model {model_id}...")
# AutoModelForCausalLM cocok untuk tugas text generation
# device_map="auto" akan mencoba menggunakan GPU jika tersedia (membutuhkan 'accelerate')
# torch_dtype="auto" mencoba menggunakan presisi yang optimal
# load_in_4bit=True (opsional, membutuhkan 'bitsandbytes') mengurangi penggunaan memori
# tapi bisa sedikit menurunkan kualitas & kecepatan. Hapus jika tidak perlu/error.
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype="auto",
    device_map="auto",
    # load_in_4bit=True # Aktifkan ini jika memori sangat terbatas
)
print("Model berhasil dimuat.")

# --- 4. Siapkan Prompt ---
# Menggunakan format chat/instruct yang disarankan untuk model Mistral Instruct
# [INST] ... [/INST]
prompt = "[INST] Ghiraldy adalah seorang mahasiswa Teknik Informatika dari Sorong. Dia sedang belajar Deep Learning. Buatkan paragraf singkat tentang proyek NLP apa yang mungkin menarik baginya! [/INST]"

print(f"\nPrompt Input:\n{prompt}\n")

# --- 5. Tokenisasi Prompt ---
# Mengubah teks menjadi token ID dan memindahkannya ke device (GPU/CPU)
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

# --- 6. Hasilkan Teks ---
print("Menghasilkan teks...")
# Menggunakan model.generate()
# max_new_tokens: Batas maksimal kata baru yang dihasilkan
# do_sample=True: Menggunakan sampling, bukan greedy search (lebih kreatif)
# temperature: Mengontrol 'keacakan' (lebih tinggi = lebih acak)
# top_p: Nucleus sampling, memilih kata dari probabilitas kumulatif teratas
# eos_token_id: Memberi tahu model kapan harus berhenti (jika menemukan token ini)
outputs = model.generate(
    **inputs,
    max_new_tokens=200,
    do_sample=True,
    temperature=0.7,
    top_p=0.9,
    eos_token_id=tokenizer.eos_token_id,
    pad_token_id=tokenizer.pad_token_id # Menentukan padding token
)

# --- 7. Dekode Hasil ---
# Mengubah token ID kembali menjadi teks
# Kita hanya ingin bagian yang baru digenerate, jadi kita skip prompt awal
generated_text = tokenizer.decode(outputs[0][inputs["input_ids"].shape[1]:], skip_special_tokens=True)

print("\n--- Teks yang Dihasilkan ---")
print(generated_text.strip())
```

```python
from transformers import pipeline

print("Memuat pipeline NER...")
# Menggunakan 'aggregation_strategy=simple' agar pipeline
# secara otomatis menggabungkan token-token yang merupakan bagian
# dari satu entitas (misal: 'Ghiraldy', 'Patrick' menjadi 'Ghiraldy Patrick').
ner_pipeline = pipeline(
    "ner",
    model="cahya/bert-base-indonesian-ner",
    aggregation_strategy="simple"
)
print("Pipeline NER siap.\n")

# Teks Contoh (dengan referensi ke Sorong dan nama Anda)
text_ner = """
Ghiraldy Patrick Toshka Anggawan adalah seorang mahasiswa dari Universitas Indonesia (UI)
yang saat ini sedang melakukan penelitian Deep Learning di kota Sorong, Papua Barat.
Dia berkolaborasi dengan PT Telekomunikasi Indonesia untuk proyeknya.
"""

print(f"Teks Input NER:\n{text_ner}")

# Lakukan NER
ner_results = ner_pipeline(text_ner)

print("\n--- Hasil Ekstraksi Entitas (NER) ---")
if ner_results:
    for entity in ner_results:
        print(f"Entitas : {entity['word']}")
        print(f"Tipe    : {entity['entity_group']}")
        print(f"Skor    : {entity['score']:.4f}")
        print("-" * 20)
else:
    print("Tidak ada entitas yang ditemukan.")
```

```python
import os
import torch
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
# --- PERUBAHAN IMPORT EMBEDDING ---
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.chains import RetrievalQA
# --- IMPORT BARU UNTUK HUGGINGFACEPIPELINE ---
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
from langchain_community.llms import HuggingFacePipeline
from langchain.prompts import PromptTemplate

# --- (Opsional) Set Hugging Face Token (masih berguna untuk download) ---
# os.environ["HUGGINGFACE_TOKEN"] = "hf_YOUR_TOKEN_HERE"

# --- Muat & Ekstrak PDF ---
pdf_path = "Laporan_NLP_Sorong.pdf" # Pastikan path ini benar!
loader = PyPDFLoader(pdf_path)
documents = loader.load()
print(f"Berhasil memuat {len(documents)} halaman dari PDF.")

# --- Pecah Teks ---
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=150)
docs_chunks = text_splitter.split_documents(documents)
print(f"Teks dipecah menjadi {len(docs_chunks)} chunks.")

# --- Muat Model Embedding (Versi Baru) ---
model_name = "sentence-transformers/all-mpnet-base-v2"
embeddings = HuggingFaceEmbeddings(model_name=model_name)
print("Model embedding dimuat.")

# --- Buat Vector Store (FAISS) ---
print("Membuat vector store FAISS...")
vectorstore = FAISS.from_documents(docs_chunks, embeddings)
print("Vector store siap.")

# --- PERUBAHAN: MUAT MODEL & TOKENIZER LOKAL UNTUK PIPELINE ---
print("Memuat model & tokenizer untuk pipeline...")
repo_id = "google/flan-t5-large"
tokenizer = AutoTokenizer.from_pretrained(repo_id)
# Gunakan AutoModelForSeq2SeqLM untuk model T5
model = AutoModelForSeq2SeqLM.from_pretrained(
    repo_id,
    torch_dtype="auto",
    device_map="auto" # Coba gunakan GPU
)

# --- PERUBAHAN: BUAT TRANSFORMERS PIPELINE ---
pipe = pipeline(
    "text2text-generation", # T5 adalah model text-to-text
    model=model,
    tokenizer=tokenizer,
    max_length=512, # Sesuaikan dengan kebutuhan
    temperature=0.1,
    top_p=0.95
)

# --- PERUBAHAN: GUNAKAN HUGGINGFACEPIPELINE ---
llm = HuggingFacePipeline(pipeline=pipe)
print("LLM (HuggingFacePipeline) dimuat.")

# --- Buat RAG Chain (RetrievalQA) ---
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

# --- PERUBAHAN: GUNAKAN .invoke() BUKAN .__call__() ---
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)
print("RAG Chain siap.")

# --- Ajukan Pertanyaan ---
query = "Apa poin utama dari dokumen ini?"
print(f"\nPertanyaan: {query}")

# --- PERUBAHAN: GUNAKAN .invoke() BUKAN .__call__() ---
# Menggunakan .invoke() sesuai DeprecationWarning
result = qa_chain.invoke({"query": query}) # Menggunakan invoke()

print("\n--- Jawaban ---")
print(result['result'])

print("\n--- Sumber Dokumen ---")
for doc in result['source_documents']:
    print(f"- Halaman {doc.metadata.get('page', '?')}: {doc.page_content[:150]}...")
```

```python
# --- 1. Instalasi Pustaka ---
# Jalankan sel ini sekali, lalu WAJIB RESTART KERNEL/SESSION!
# !pip install --upgrade langchain langchain-core langchain-community langchain-huggingface pypdf faiss-cpu sentence-transformers transformers torch accelerate bitsandbytes

print("Pastikan Anda sudah menjalankan instalasi di atas dan MERESTART KERNEL/SESSION!")

# --- 2. Import Pustaka yang Diperlukan ---
import os
import torch
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.chains import RetrievalQA
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
from langchain_community.llms import HuggingFacePipeline
import warnings

# Abaikan peringatan yang tidak krusial untuk demo ini
warnings.filterwarnings("ignore")

# --- 3. Konfigurasi Awal ---

# GANTI INI DENGAN PATH PDF ANDA!
# Pastikan file Laporan_NLP_Sorong.pdf sudah diunggah/ada.
pdf_path = "Laporan_NLP_Sorong.pdf"

# Nama model embedding dari Hugging Face
embedding_model_name = "sentence-transformers/all-MiniLM-L6-v2"

# Nama model LLM dari Hugging Face (Flan-T5-Base cukup kecil)
llm_repo_id = "google/flan-t5-base"

# Cek ketersediaan GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Menggunakan device: {device}")

# --- 4. Muat dan Pecah PDF ---
try:
    print(f"Memuat PDF dari: {pdf_path}...")
    loader = PyPDFLoader(pdf_path)
    documents = loader.load()
    print(f"Berhasil memuat {len(documents)} halaman.")

    print("Memecah teks menjadi chunks...")
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=150)
    docs_chunks = text_splitter.split_documents(documents)
    print(f"Teks dipecah menjadi {len(docs_chunks)} chunks.")

except Exception as e:
    print(f"Error saat memuat atau memecah PDF: {e}")
    print("Pastikan path PDF sudah benar dan file tidak rusak.")
    exit() # Hentikan jika gagal

# --- 5. Buat Embeddings & Vector Store ---
try:
    print(f"Memuat model embedding: {embedding_model_name}...")
    embeddings = HuggingFaceEmbeddings(
        model_name=embedding_model_name,
        model_kwargs={'device': device} # Gunakan GPU jika ada
    )
    print("Model embedding dimuat.")

    print("Membuat vector store FAISS...")
    vectorstore = FAISS.from_documents(docs_chunks, embeddings)
    print("Vector store siap.")

except Exception as e:
    print(f"Error saat membuat embedding atau vector store: {e}")
    exit()

# --- 6. Muat LLM via HuggingFacePipeline ---
try:
    print(f"Memuat tokenizer dan model LLM: {llm_repo_id}...")
    tokenizer = AutoTokenizer.from_pretrained(llm_repo_id)
    # Gunakan AutoModelForSeq2SeqLM untuk model T5
    model = AutoModelForSeq2SeqLM.from_pretrained(
        llm_repo_id,
        torch_dtype=torch.float16 if device == "cuda" else torch.float32, # Gunakan float16 di GPU
        device_map="auto" # Otomatis pindah ke GPU jika ada
    )
    print("Model LLM dimuat.")

    # Buat pipeline transformers
    pipe = pipeline(
        "text2text-generation",
        model=model,
        tokenizer=tokenizer,
        max_new_tokens=512, # Batasi token baru
        temperature=0.2, # Suhu rendah agar lebih faktual
        top_p=0.95
    )

    # Bungkus dengan LangChain HuggingFacePipeline
    llm = HuggingFacePipeline(pipeline=pipe)
    print("HuggingFacePipeline siap.")

except Exception as e:
    print(f"Error saat memuat LLM atau pipeline: {e}")
    print("Model ini mungkin membutuhkan banyak RAM/VRAM. Coba gunakan runtime GPU di Colab.")
    exit()

# --- 7. Buat RAG Chain (RetrievalQA) ---
# Siapkan retriever untuk mengambil dokumen relevan
retriever = vectorstore.as_retriever(search_kwargs={"k": 3}) # Ambil 3 chunk teratas

# Buat RAG chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff", # Tipe chain sederhana: masukkan semua chunk ke prompt
    retriever=retriever,
    return_source_documents=True
)
print("RAG Chain siap.")

# --- 8. Interaksi dengan Chatbot ---
def jawab_pertanyaan(pertanyaan):
    """Fungsi untuk mengajukan pertanyaan dan mendapatkan jawaban."""
    print(f"\n🤔 Pertanyaan: {pertanyaan}")
    try:
        # Gunakan .invoke() untuk LangChain versi baru
        result = qa_chain.invoke({"query": pertanyaan})
        print("\n💡 Jawaban:")
        print(result['result'])
        # print("\n📚 Sumber:")
        # for doc in result['source_documents']:
        #     print(f"  - Halaman {doc.metadata.get('page', '?')}: {doc.page_content[:100]}...")
    except Exception as e:
        print(f"Error saat menjalankan RAG chain: {e}")

# Ajukan beberapa pertanyaan
jawab_pertanyaan("Apa itu NLP?")
jawab_pertanyaan("Di mana Sorong berada?")
jawab_pertanyaan("Apa tantangan utama NLP di Indonesia Timur?")
jawab_pertanyaan("Siapa Ghiraldy Patrick?")
```
